```python
import pandas as pd
```

```python
df = pd.read_csv("sub_train_wo_unclassified_prepost.csv")
df
```

```python
no_damage = df[df["bldg_damage_code"] == 0][0:10000]
```

```python
no_damage
```

```python
minor_damage = df[df["bldg_damage_code"] == 1][0:10000]
minor_damage
```

```python
major_damage = df[df["bldg_damage_code"] == 2][0:10000]
major_damage
```

```python
destroyed = df[df["bldg_damage_code"] == 3][0:10000]
destroyed
```

```python
combined_df = no_damage.append(minor_damage).append(major_damage).append(destroyed)
result_df = combined_df.sample(frac=1, random_state=1)
result_df
```

```python
result_df.to_csv('sub_train_wo_unclassified_prepost_10000_each_type.csv',index=0)
```
